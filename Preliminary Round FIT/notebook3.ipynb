```python
features = df.drop(['happiness_score', 'traffic_density'], axis=1)
target = df['happiness_score']
```

```python
from sklearn.preprocessing import RobustScaler

features = ['total_area (km2)', 'population', 'densities', 'traffic_density_encoded', 'green_open_space',
            'hdi', 'gross_regional_domestic_product', 'total_landfills', 'solid_waste_generated']
scaler = RobustScaler()
X = scaler.fit_transform(df[features])
dfx = pd.DataFrame(data=X, columns=features)
dfx.head()
```

```python
dfx = pd.DataFrame(data=dfx, columns=features)
dfx['happiness_score'] = target.values
dfx.head()
```

```python
dfx.describe()
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVR
from sklearn.ensemble import GradientBoostingRegressor, AdaBoostRegressor, RandomForestRegressor
from xgboost import XGBRFRegressor
```

```python
X = dfx.drop(['happiness_score'], axis=1)
y = dfx['happiness_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
from sklearn.metrics import mean_squared_error

model = SVR(C=1, epsilon=0.5, gamma=1)

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

mse
```

```python
#GradientBoostingRegressor, AdaBoostRegressor, RandomForestRegressor from xgboost import XGBRFRegressor
model_params = {
    'GradientBoostingRegressor': {
        'model': GradientBoostingRegressor(),
        'params' : {
            'n_estimators': [10,20,30],
            'learning_rate': [0.01, 0.05, 0.1, 0.2],
            'max_depth': [3, 4, 5, 6],
            'subsample': [0.7, 0.8, 0.9, 1.0],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }  
    },
    'AdaBoostReg': {
        'model': AdaBoostRegressor(),
        'params' : {
            'n_estimators': [10,20,30],
            'learning_rate': [0.01, 0.1, 0.5, 1.0],
            'loss': ['linear', 'square', 'exponential']
        }
    },
    'Random Forest' : {
        'model': RandomForestRegressor(),
        'params': {
            'n_estimators': [10,20,30],
            'max_depth': [None, 10, 20, 30],
            'min_samples_split': [2, 5, 10],
            'min_samples_leaf': [1, 2, 4]
        }
    },
    'XGBoost' : {
        'model': XGBRFRegressor(),
        'params': {
            'n_estimators': [10,20,30],
            'learning_rate': [0.01, 0.1, 0.2],
            'subsample': [0.7, 0.8, 0.9],
            'colsample_bynode': [0.7, 0.8, 0.9]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv = 6, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
print(clf.best_estimator_)
```

```python
df_score = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_score
```

```python
best_model = clf.best_estimator_

y_pred = best_model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

mse
```

```python
model = AdaBoostRegressor(learning_rate=0.1, loss='exponential', n_estimators= 10)

model.fit(X_train, y_train)

y_pred_ada = model.predict(X_test)

print(mean_squared_error(y_test, y_pred_ada))
```

```python
model2 = RandomForestRegressor(max_depth=30, min_samples_leaf=2, min_samples_split=5, n_estimators=20)

model2.fit(X_train, y_train)

y_pred_rf = model2.predict(X_test)

print(mean_squared_error(y_test, y_pred_rf))
```

```python
#GradientBoostingRegressor,0.130149124281588,"{'learning_rate': 0.1, 'max_depth': 4, 'min_samples_leaf': 2, 'min_samples_split': 5, 'n_estimators': 30, 'subsample': 0.7}"

gb_model = GradientBoostingRegressor(learning_rate= 0.1, max_depth = 4, min_samples_leaf = 2, min_samples_split = 5, n_estimators = 30, subsample = 0.7)
gb_model.fit(X_train, y_train)
y_pred_ridge = gb_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred_ridge)
mse
```

```python
from sklearn.feature_selection import RFE

svr = SVR(C=1, epsilon=0.5, gamma=1)
xgb = XGBRFRegressor(colsample_bynode=0.9, learning_rate=0.2, n_estimators=10, subsample=0.8)

rfe = RFE(estimator=xgb, n_features_to_select=9, step=1)

# Fit RFE
rfe.fit(X_train, y_train)

# Get selected features
selected_features = X.columns[rfe.support_]

# Use selected features for further modeling
X_train_selected = rfe.transform(X_train)
X_test_selected = rfe.transform(X_test)
```

```python
# Get feature importances from the estimator (RandomForestRegressor in this case)
feature_importances = rfe.estimator_.feature_importances_

# Sort indices by importance
sorted_idx = feature_importances.argsort()

# Plot
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_idx)), feature_importances[sorted_idx], align='center')
plt.yticks(range(len(sorted_idx)), X_train.columns[sorted_idx])
plt.xlabel('Feature Importance')
plt.title('Feature Importances (RFE with RandomForestRegressor)')
plt.show()
```
