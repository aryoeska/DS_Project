```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./train.csv')
df.head()
```

```python
df.info()
```

```python
obj_features = ['total_area (km2)', 'population', 'green_open_space', 'gross_regional_domestic_product', 'solid_waste_generated']

for col in obj_features:
    df[col] = df[col].replace('#REF!', np.nan)
    df[col] = df[col].astype(str)
    df[col] = df[col].str.replace(',','').astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
df['green_open_space'].value_counts()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['green_open_space'], bins=10)
```

```python
# missing_features = ['green_open_space', 'total_landfills', 'solid_waste_generated']

df['green_open_space'].fillna(df['green_open_space'].mean(), inplace=True)
df['total_landfills'].fillna(df['total_landfills'].mean(), inplace=True)
df['solid_waste_generated'].fillna(df['solid_waste_generated'].median(), inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder()

df['traffic_density_encoded'] = ordinal.fit_transform(df[['traffic_density']])
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVR
from sklearn.linear_model import LinearRegression, RANSACRegressor
```

```python
X = df.drop(['id', 'city_or_regency', 'traffic_density', 'happiness_score'], axis=1)
y = df['happiness_score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
model_params = {
    'svc': {
        'model': SVR(),
        'params' : {
            'C': [0.1, 1, 10],
            'gamma': [0.1, 1, 'scale'],
            'epsilon': [0.1, 0.2, 0.5]
        }  
    },
    'RANSAC': {
        'model': RANSACRegressor(),
        'params' : {
            'residual_threshold': [0.1, 1.0, 10.0],
            'max_trials': [100, 500, 1000]
        }
    },
    'linear_regression' : {
        'model': LinearRegression(),
        'params': {
            'n_jobs': [None, -1, 1, 2]
        }
    }
}
```

```python
scores = []

for model_name, mp in model_params.items():
    clf = GridSearchCV(mp['model'], mp['params'], cv = 10, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model':model_name,
        'best_score':clf.best_score_,
        'best_params':clf.best_params_
    })
```

```python
df_score = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_score
```

```python
from sklearn.metrics import mean_squared_error

model = SVR(C=1, epsilon=0.5, gamma='scale')

model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)

mse
```

```python
model3 = RANSACRegressor(max_trials=1000, residual_threshold=10)

model3.fit(X_train,y_train)

y_pred2 = model3.predict(X_test)

mse = mean_squared_error(y_test, y_pred2)

mse
```

```python
model2 = LinearRegression(n_jobs=None)

model2.fit(X_train,y_train)

y_pred2 = model2.predict(X_test)

mse = mean_squared_error(y_test, y_pred2)

mse
```
